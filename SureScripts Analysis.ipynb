```python
# Load typed data from csv

import pandas as pd
import numpy as np
from IPython.display import HTML
pd.set_option('display.max_rows', 500)


def html(res, index=False):
    return HTML(res.to_html(index=index))

print pd.__version__


```

```python
dispenses = pd.read_csv("../fixed_drug_details.csv", 
    sep="|", 
    dtype={'DrugCodedProductCode': 'string'}, 
    parse_dates=[9, 16] )

pts = pd.read_csv("../Medication_History_Records.csv",
    sep=",",
    parse_dates=[3])

pts.dtypes

joined = dispenses.merge(pts)
backwards = pts.merge(dispenses[['messageid']].drop_duplicates())
```

```python
deduplicated = joined[['patientid', 'DrugCodedProductCode','DateWritten', 'FillDate', 'RefillQuantity',  'FillNumber',  'DaySupply',  'QuantityValue']].sort(['patientid', 'DrugCodedProductCode','FillDate']).drop_duplicates()
table_rxs = deduplicated[['patientid', 'DrugCodedProductCode','DateWritten', 'RefillQuantity']].drop_duplicates().reset_index()
table_rxs['id'] = pd.Series(data=table_rxs.index)

table_fills = deduplicated.merge(table_rxs)[['id','FillDate', 'FillNumber', 'DaySupply', 'QuantityValue']].drop_duplicates().reset_index().rename(columns={
 'id':'rx_id',
 'FillDate': 'date',
 'FillNumber': 'nominal_fill_number',
 'DaySupply': 'days_supply',
 'QuantityValue': 'quantity'
})[['rx_id', 'date', 'nominal_fill_number', 'days_supply', 'quantity']]
table_fills['id'] = pd.Series(data=table_fills.index)

table_rxs = table_rxs.rename(columns={
  'patientid': 'patient_id',
  'DrugCodedProductCode':'ndc',
  'DateWritten':'date',
  'RefillQuantity':'refills'
})[['id', 'patient_id', 'ndc', 'date', 'refills']]


```

```python
table_pts = joined[['patientid', 'MessageDate']].sort(['patientid']).drop_duplicates().rename(columns={
  'patientid': 'id',
  'MessageDate':'sync_date'
}).groupby('id').agg({'sync_date':'max'}).reset_index()
```

```python
import sqlite3
conn = sqlite3.connect('rxnorm/py-umls/databases/rxnorm.db')

print table_rxs.dtypes
pd.io.sql.to_sql(table_rxs, 'rxs', conn, if_exists='replace', index=False)
pd.io.sql.to_sql(table_fills, 'fills', conn, if_exists='replace', index=False)
pd.io.sql.to_sql(table_pts, 'pts', conn, if_exists='replace', index=False)

```

```python
expected_vs_actual = pd.io.sql.read_sql("""
WITH fill_counts as (
  SELECT fills.rx_id as rx_id, count(*) as actual_fills
    from rxs
    join fills on rxs.id=fills.rx_id
 --   join drug_treatments_by_ndc t on t.ndc=rxs.ndc
    join pts on pts.id = rxs.patient_id
  --  where t.treatment_intent='Hypertension'
  --  and refills = 11
    and date(pts.sync_date) > date(rxs.date,'+' || rxs.refills || ' month')
    group by fills.rx_id) 
SELECT refills+1 as prescribed_fills, actual_fills as dispensed_fills, count(*) as count
  from fill_counts c
  join rxs r on c.rx_id=r.id
  group by r.refills, actual_fills
  having r.refills >= 5 and count(*) > 25
  order by r.refills asc, actual_fills asc
""", conn)
```

```python
%matplotlib inline
from ggplot import *

ggplot(expected_vs_actual, aes('dispensed_fills', y='count')) + \
    geom_bar(color='black', stat='identity')  + facet_wrap("prescribed_fills", scales="fixed") + \
    xlab("# Actual dispenses") + ylab("Count") 
```

```python
import sqlite3
conn = sqlite3.connect('rxnorm/rxnorm.sqlite')

ndcq =""" select a.ATV as NDC, c.rxcui as RXCUI, c.tty as TTY, str as NAME
  from rxnsat a join rxnconso c on a.rxaui=c.rxaui
  where a.ATN='NDC' and a.SAB='RXNORM'"""

ndc_map = pd.io.sql.read_sql(ndcq, conn)
ndc_map[:10]

names_by_rxcui = ndc_map[['RXCUI', 'TTY', 'NAME']].drop_duplicates()
```

```python
top_rxcuis = dispenses.merge(ndc_map,  left_on='DrugCodedProductCode', right_on='NDC').\
            RXCUI.value_counts().head(100)
    
res = pd.DataFrame({'count': top_rxcuis}).merge(names_by_rxcui, left_index=True, right_on=['RXCUI'])
html(res)
```
