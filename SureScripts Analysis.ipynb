## Just exploring the SureScripts dataset...


```python
# Load typed data from csv

import pandas as pd
import numpy as np
from IPython.display import HTML
pd.set_option('display.max_rows', 500)


def html(res, index=False):
    return HTML(res.to_html(index=index))


```

```python
dispenses = pd.read_csv("../fixed_drug_details.csv", 
    sep="|", 
    dtype={'DrugCodedProductCode': 'string'}, 
    parse_dates=[9, 16] )

dispenses.dtypes
```

```python
## Is the average quantity of drug supplied on a prescription
## increasing over time? Let's see...

grouped = dispenses.groupby(lambda x: int(x/1000)).agg(np.mean)
grouped['rank'] = pd.Series(data=grouped.index)

#dispenses['rx_order'] = dispenses.DateWritten.rank(method='dense')
#bydate = dispenses.sort('DateWritten').concat(range(len(axis=1)
#somed = dispenses.sort('DateWritten')[0:10000].reset_index()
```

On a quick visual examination, it doesn't look like there's really an effect.

(Although this is what stats are for.)

```python
%matplotlib inline
from ggplot import *

ggplot(pd.ewma(grouped, 10), aes('rank', 'DaySupply', xmin=0, xlim=1000)) + \
    geom_line(color='black') + xlim(0, 1000)
```

```python
top_ndcs = dispenses.DrugCodedProductCode.value_counts().head(100)
names_by_ndc = dispenses[['DrugCodedProductCode', 'DrugDescription']].drop_duplicates()

res = pd.DataFrame({'count': top_ndcs}).merge(names_by_ndc, left_index=True, right_on=['DrugCodedProductCode'])
html(res)
```

```python
# Most common drugs by free-text name?

dispenses.groupby('DrugDescription').count().sort('messageid', ascending=False).messageid.head(10)
```

```python
import sqlite3
conn = sqlite3.connect('rxnorm/rxnorm.sqlite')

ndcq =""" select a.ATV as NDC, c.rxcui as RXCUI, c.tty as TTY, str as NAME
  from rxnsat a join rxnconso c on a.rxaui=c.rxaui
  where a.ATN='NDC' and a.SAB='RXNORM'"""

ndc_map = pd.io.sql.read_sql(ndcq, conn)
ndc_map[:10]

names_by_rxcui = ndc_map[['RXCUI', 'TTY', 'NAME']].drop_duplicates()
```

```python
top_rxcuis = dispenses.merge(ndc_map,  left_on='DrugCodedProductCode', right_on='NDC').\
            RXCUI.value_counts().head(100)
    
res = pd.DataFrame({'count': top_rxcuis}).merge(names_by_rxcui, left_index=True, right_on=['RXCUI'])
html(res)
```
